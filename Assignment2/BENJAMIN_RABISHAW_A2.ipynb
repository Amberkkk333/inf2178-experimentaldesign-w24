```python
# Data Cleaning

## Removing columns for Research Question 1
dfRemove = ['LOC_ID', 'LOC_NAME', 'ADDRESS', 'PCODE', 'bldg_type', 'ward', 'BLDGNAME']

## Making new dataframe for Research Question 1
dfc = df.drop(dfRemove, axis='columns')

## Setting index for dataframe
dfc.set_index('_id', inplace=True)

## Grouping by Auspice for Research Question 1
##dfc = dfc.groupby('AUSPICE')

## Surveying first rows of cleaned dataframe
dfc.head(25)
```

```python
## Exploring dataframe using a boxplot

ax = sns.boxplot(x='AUSPICE', y='TOTSPACE', data=dfc, color='#99c2a2')
ax = sns.swarmplot(x="AUSPICE", y="TOTSPACE", data=dfc, color='#7d0013')
plt.show()
```

```python
# Assumption 1: Testing the normality-of-residuals assumption for ANOVA using a Shapiro-Wilk test

## Ordinary Least Squares (OLS) model
model = sm.OLS.from_formula('TOTSPACE ~ C(AUSPICE)', data=dfc)
result = model.fit()
residuals = result.resid

### Performing Shapiro-Wilk test for normality of residuals
shapiro_statistic, shapiro_p_value = shapiro(residuals)

print("Shapiro-Wilk Test Statistic:", shapiro_statistic)
print("p-value:", shapiro_p_value)

## Transforming the TOTSPACE data using the Box-Cox method
bcTOTSPACE, lambda_value = stats.boxcox(dfc['TOTSPACE'])

### Adding the transformed TOTSPACE data as a column in the main dataframe for later use
dfc['bcTOTSPACE'] = bcTOTSPACE

## Re-making OLS model and re-running Shapiro-Wilk test
model2 = sm.OLS.from_formula('bcTOTSPACE ~ C(AUSPICE)', data=dfc)
result2 = model2.fit()
residuals2 = result2.resid

shapiro_statistic2, shapiro_p_value2 = shapiro(residuals2)

print("Shapiro-Wilk Test Statistic for transformed data:", shapiro_statistic2)
print("p-value for transformed data:", shapiro_p_value2)
```

```python
# Assumption 2: Testing the homogeneity of variances for ANOVA using a Brown-Forsythe test, a variant of Levene's test

## Performing the initial Brown-Forsythe test
bf_test_statistic, bf_p_value = levene(*[group['bcTOTSPACE'] for _, group in dfc.groupby('AUSPICE')], center='median')
print("Brown-Forsythe Test Statistic:", bf_test_statistic)
print("p-value:", bf_p_value)
print()

## Transforming the Box-Cox-transformed TOTSPACE values further by taking their square root
sqrtbcTOTSPACE = np.sqrt(bcTOTSPACE)
dfc['sqrtbcTOTSPACE'] = sqrtbcTOTSPACE

# Performing Brown-Forsythe test again on the additionally transformed data
bf_test_statistic, bf_p_value = levene(*[group['sqrtbcTOTSPACE'] for _, group in dfc.groupby('AUSPICE')], center='median')

print("Brown-Forsythe Test Statistic for transformed data:", bf_test_statistic)
print("p-value for transformed data:", bf_p_value)
```

```python
# One-way ANOVA for Research Question 1

## Ordinary Least Squares (OLS) model
RQ1model = ols('sqrtbcTOTSPACE ~ C(AUSPICE)', data=dfc).fit()

## Displaying ANOVA table
anova_table = sm.stats.anova_lm(RQ1model, typ=2)
print(anova_table)

```

```python
# Post-hoc Tukey test for Research Question 1

## Performing the Tukey test
tukey_result = pairwise_tukeyhsd(dfc['sqrtbcTOTSPACE'], dfc['AUSPICE'])

## Displaying test results
print(tukey_result.summary())
```
