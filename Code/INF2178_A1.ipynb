```python
# Load data and head data
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'INF2178_A1_data.xlsx'
data = pd.read_excel(file_path)
data.head()
```

```python
# Clean data fill missing value
data.fillna(data.mean(), inplace=True)
data.fillna(data.mode().iloc[0], inplace=True)

data.head()
```

```python
# Summary statistics for numerical columns
print("Summary Statistics:\n", data.describe())
```

```python
# Box plot for 'SERVICE_USER_COUNT'
plt.figure(figsize=(8, 5))
plt.boxplot(data['SERVICE_USER_COUNT'])
plt.title('Box Plot of Service User Count')
plt.ylabel('Service User Count')
plt.show()
```

```python
# Plotting the distribution of 'SERVICE_USER_COUNT'
plt.hist(data['SERVICE_USER_COUNT'], bins=30, edgecolor='black')
plt.title('Distribution of Service User Count')
plt.xlabel('Service User Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a scatter plot of the relationship between 'SERVICE_USER_COUNT' and 'CAPACITY_ACTUAL_BED'
plt.scatter(data['SERVICE_USER_COUNT'], data['CAPACITY_ACTUAL_BED'])
plt.title('Service User Count & Capacity Actual Bed')
plt.xlabel('Service User Count')
plt.ylabel('Capacity Actual Bed')
plt.show()
```

```python
# T-test and P-value
from scipy.stats import ttest_ind

room_based = data[data['CAPACITY_TYPE'] == 'Room Based Capacity']['SERVICE_USER_COUNT']
bed_based = data[data['CAPACITY_TYPE'] == 'Bed Based Capacity']['SERVICE_USER_COUNT']

t_stat, p_value = ttest_ind(room_based, bed_based, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
